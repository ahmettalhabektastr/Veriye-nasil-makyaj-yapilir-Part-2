> ## **Gerekli kütüphaneleri import edelim**

```python
import matplotlib.pyplot as plt
#Matplotlib kütüphanesini "plt" olarak kısaltıp import ettik
%matplotlib inline

import seaborn as sns
#Seaborn kütüphanesini "sns" diye kısaltıp import ettik
```

**Seaborn** kütüphanesinin içinde yüklü bazı veri setleri var onları görmek için şu kodu kullanıyoruz:

```python
sns.get_dataset_names()
```

**Seaborn** kütüphanesinin titanic veri setini kullanıcam

```python
titanic=sns.load_dataset("titanic")
```

Eksik değerler ile nasıl başa çıkılcağını henüz göstermedin o yüzden eksik yaş verilerini kullanmıcam.

```python
titanic_full_age=titanic[titanic["age"].notnull()]
```

```python
titanic_full_age.isnull().sum()
```

> # **lineplot**

Hem kısa hem de uzun zaman dilimlerindeki değişiklikleri izlemek için **çizgi grafikleri (line plot)** kullanılır. **Çizgi grafikleri**, daha küçük değişiklikler olduğunda **çubuk grafiklerinden** daha iyi performans gösterir. **Çizgi grafikleri**, aynı zaman diliminde birden çok gruptaki değişiklikleri karşılaştırmak için kullanılabilir.

```python
sns.lineplot(data=titanic_full_age[["age","fare"]])
#Bir çizgi grafiği oluştur ve veri setimiz bu olsun: "titanic_full_age" 
#veri seitnin yalnızca iki sütununu verdim böylece otomatikmen bu verileri x eksenine koyucak
```

```python
sns.lineplot(data=titanic_full_age,x="age",y="fare");
#verilerimiz şuradan: "titanic_full_age"
#"age" sütununu x eksenine
#"fare" sütununu y eksenine koy

#eğer kodun sonunda noktalı virgül (";") koyarsanız, karşınıza şunun gibi bir ifade gelmicektir:
#"<AxesSubplot:xlabel='age', ylabel='fare'>"
```

**Seaborn**'de sabit stiller vardır. Bunlar figürün arka planını ayarlar. yaygın olarak kullanılan bazı stiller şunlardır: *darkgrid*, *whitegrid*, *dark*, *white* ve *ticks*'dir.

```python
sns.set_style("darkgrid")
```

```python
sns.lineplot(data=titanic_full_age,x="age",y="fare");
```

```python
sns.set_style("whitegrid")
```

```python
sns.lineplot(data=titanic_full_age[["age","fare"]]);
```

```python
sns.set_style("white")
#Arkaplanı beyaz yani default duruma alır.
```

> # **scatterplot**

İki değişkenin aralarındaki korelasyon güzelse, **dağılım grafiği ( scatter plot)** kullanmanız güzel olur.

```python
sns.scatterplot(x="age",y="fare",hue="survived",data=titanic_full_age);
#dağılım grafiği oluştur ve bu değerleri "survived" sütununa göre karşılaştır
```

> # **histplot**

Yaygın bir grafik aracı **histogramdır**. Veri dağılımını gözler önüne sermek için birebirdir.

```python
sns.histplot(data=titanic_full_age, x="who");
#Bir histogram oluştur ve "who" sütununun değerlerini x eksenine koy
```

```python
sns.histplot(data=titanic_full_age, y="who", fill=False);
#"who" sütununun değerlerini y eksenine koy
#Dikdörtgenlerin içini doldurma!
```

```python
sns.histplot(data=titanic_full_age, x="who",kde=True);
#kde=kernel density estimate, yoğunluk kestirimi 
#"kde" sayesinde verinin dağılımını rahatça görebiliyoruz
```

```python
sns.histplot(data=titanic_full_age, x="pclass", hue="survived");
#verdiğimiz değerleri bir sütuna göre de karşılaştırabiliyoruz
```

Mavi ve turuncu üst üste geldiğinde siyah görüyoruz. Bu durumu düzeltmek için **multiple = "stack"** kullanabilirsiniz.

```python
sns.histplot(data=titanic_full_age, x="pclass", hue="survived",multiple="stack");
```

Yada **element = "step"** kullanabilirsiniz.

```python
sns.histplot(data=titanic_full_age, x="pclass", hue="survived", element="step");
```

> # **displot**

**displot**, **histplot**'a çok benziyor.

```python
sns.displot(titanic_full_age.age,bins=10,kde=True);
#bins=n demek bize n tane dikdörtgen göster demektir.
```

```python
sns.displot(titanic_full_age.age,bins=50,kde=True);
```

> # **boxplot**

**Kutu grafikleri (Box plot)**, özellikle çeşitli gruplar arasında karşılaştırma yaparken, sayısal veri değerlerinin dağılımlarını göstermek için kullanılır. Bir bakışta üst düzey bilgi vermek ve bir veri kümesinin, *varyansı* ve *aykırı değerleri* gibi ayrıntıları sağlamak için tasarlanmıştır.

```python
sns.boxplot(x="age",y="sex",data=titanic_full_age,hue="survived");
#Bir kutu grafiği oluşturuluyor
```

> # **countplot**

Çubukları kullanarak her bir kategori kutusundaki gözlem sayısını görüntülemek için **sütun grafiği (count plot)** yöntemini kullanılır.

```python
sns.countplot(x=titanic_full_age.embarked);
#kendisi direk toplamı hesaplıyor,ayrıca value_counts() yazmamalısnız.
```

```python
sns.countplot(y=titanic_full_age.embarked);
#data= diye belirtmeden direk veriyi verebiliriz
```

```python
sns.countplot(x=titanic_full_age.embarked,hue=(titanic_full_age["survived"]));
```

> # **barplot** 

**Barplot** kullanarak çeşitli gruplar arasında birkaç veri setini hızlı bir şekilde karşılaştırabilirsiniz.

```python
sns.barplot(data=titanic_full_age, x="pclass", y="age");
```

```python
sns.barplot(data=titanic_full_age, x="pclass", y="age",hue="survived");
```

># **heatmap**  

**heatmap (Isı haritası)**, veri bilimcilerin kullandığı çok önemli bir grafik türüdür.

```python
sns.heatmap(titanic_full_age.corr(), annot = True);
#Bu grafiğin içine veri kümemizin korelasyonunun veriyoruz
#annot=True, karelerin içinde sayıları göster demek
```

```python
sns.heatmap(titanic_full_age.corr(), annot = False,cmap="crest");
#annot=False, karelerin içinde sayıları gösterme demek
#cmap=colour map, renk haritası
```

> # **pairplot**  

**pairplot** her sütünun her sütun ile ilişkisini çizen grafik türüdür.

```python
sns.pairplot(titanic, hue="survived")
plt.savefig("titanic.png",dpi=300)
#matplotlib sayesinde grafiği kaydediyoruz.
```

```python
sns.pairplot(titanic, hue="survived", kind="kde");
#pairplot otomatik olarak scatter plot çizer bunu değiştirmek için kind kullanıyoruz
#kind'da şu ifadeleri kullanabilirsiniz: 'scatter', 'kde', 'hist' ve 'reg']

plt.savefig("titanic_kde.png",dpi=300)
```
